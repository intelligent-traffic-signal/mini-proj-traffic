```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Sarakki_outputs/Test_weib/ncpu=1_15k/test_green10_3rdtime_run1.csv')
```

```python
data_x = data.step_time
data_y = data.total_stopped
```

```python
plt.plot(data_x, data_y)
plt.xlabel('Time Step')
plt.ylabel('Queue Length')
fig = plt.gcf()
fig.set_size_inches(10, 6)
fig.savefig('uniform', dpi=96)
```
